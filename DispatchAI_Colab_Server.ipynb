```python
# 2. Load Models (The Brain Cells)
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
from peft import PeftModel
from faster_whisper import WhisperModel
import opensmile
import os

print("â³ Loading Whisper STT...")
stt_model = WhisperModel("large-v3", device="cuda", compute_type="float16")

print("â³ Loading OpenSMILE...")
smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPSv02,
    feature_level=opensmile.FeatureLevel.Functionals,
)

print("â³ Loading Mistral 7B + 911 Adapter...")
# Base Model
base_model_id = "mistralai/Mistral-7B-v0.1"
adapter_id = "spikecodes/ai-911-operator"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16
)

model = AutoModelForCausalLM.from_pretrained(
    base_model_id,
    quantization_config=bnb_config,
    device_map="auto"
)
model = PeftModel.from_pretrained(model, adapter_id)
tokenizer = AutoTokenizer.from_pretrained(base_model_id)
tokenizer.pad_token = tokenizer.eos_token

print("âœ… All Systems Online. GPU Ready.")
```

```python
# 3. The Server Logic (With SSH Auto-Yes)
from fastapi import FastAPI, UploadFile, File
from fastapi.responses import JSONResponse
import uvicorn
import nest_asyncio
import shutil
import json
import time
import subprocess
from threading import Thread

app = FastAPI()
nest_asyncio.apply()

@app.post("/process_emergency")
async def process_emergency(audio: UploadFile = File(...)):
    start_time = time.time()

    # A. Save Audio
    temp_filename = f"temp_{audio.filename}"
    with open(temp_filename, "wb") as buffer:
        shutil.copyfileobj(audio.file, buffer)

    # B. Transcribe (Whisper)
    print("ğŸ”Š Transcribing...")
    segments, info = stt_model.transcribe(temp_filename, beam_size=5)
    transcript = " ".join([segment.text for segment in segments])
    print(f"ğŸ—£ï¸ Transcript: {transcript}")

    # C. Acoustic Analysis (OpenSMILE)
    # We look for high pitch/loudness as indicators of Panic
    features = smile.process_file(temp_filename)
    pitch_mean = features['F0semitoneFrom27.5Hz_sma3nz_amean'].values[0]
    loudness_mean = features['loudness_sma3_amean'].values[0]

    emotion_tag = "Calm"
    if pitch_mean > 35 or loudness_mean > 2.0:
        emotion_tag = "Panic"
    print(f"ğŸ’“ Acoustic Emotion: {emotion_tag}")

    # D. LLM Response
    print("ğŸ§  Thinking...")
    prompt = f"911 Operator: 9-1-1, what's your emergency?\nCaller: {transcript}\n911 Operator:"
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
    outputs = model.generate(**inputs, max_new_tokens=50, pad_token_id=tokenizer.eos_token_id)
    response_text = tokenizer.decode(outputs[0], skip_special_tokens=True).split("911 Operator:")[-1].strip()

    print(f"ğŸ¤– Response: {response_text}")

    return JSONResponse(content={
        "transcript": transcript,
        "response": response_text,
        "acoustic_emotion": emotion_tag,
        "processing_time": time.time() - start_time
    })

print("ğŸš€ Starting Tunnel...")
def start_tunnel():
    # Added StrictHostKeyChecking=no to prevent hanging on 'Are you sure?' prompt
    process = subprocess.Popen(
        ["ssh", "-o", "StrictHostKeyChecking=no", "-p", "443", "-R0:localhost:8000", "a.pinggy.io"],
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )
    for line in process.stdout:
        if "pinggy.link" in line:
            print(f"\nğŸ”— CONNECT TO THIS URL: {line.strip()}\n")

Thread(target=start_tunnel, daemon=True).start()
time.sleep(3)

print("ğŸš€ Starting Server...")
config = uvicorn.Config(app, host="0.0.0.0", port=8000)
server = uvicorn.Server(config)
await server.serve()
```
